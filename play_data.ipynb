```python
qty = len(idx_list)
for i in tqdm(range(qty), leave=True):
    idx_used = idx_list[i] #np.random.choice(idx_list)
    with fits.open('./data/broadband_'+str(idx_used)+'.fits') as hdul:
        data = hdul[0].data  # assuming the data you want is in the primary HDU
    # plt.subplot(int(np.sqrt(qty)),int(np.sqrt(qty)),i+1)
    plt.imshow(data[0], norm='log')
    plt.xticks([])
    plt.yticks([])
    plt.savefig('./data_image/broadband_'+str(idx_used))
```
